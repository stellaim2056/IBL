# The "Brain-wide Map" project

The IBL has been working on the "brain-wide map", a project that aims to map neural activity at single-spike cellular resolution across the whole mouse brain during a decision-making task.

## Recording plan

We recorded activity using Neuropixel probes that were targetted in the brain following a grid system comprising insertions seperated by 500 um.

Our recording system aimed to:
- Obtain uniform coverage of the brain
- Maximise recordings from areas that share anatomical connections (for this reason, the cerebellum and medulla were sampled from the right hemisphere instead of the left)
- Replicate each recording site in at least two laboratories
- Assess the reproducibility of recordings across subjects and laboratories (via the "repeated site")
- Assess bilateral, interhemispheric interactions (via the "bilateral sites")

# Overview of the experiment

## Standardised training pipeline

All data released as part of the "brain-wide map" dataset were recorded from mice that went through a standardised training pipeline developed at the IBL. Following headbar implantation mice are trained on a standardised behavioral rig until they expert are at the task. They are then moved over to the electrohphysiology rig and once habituated, Neuropixel recordings are performed. To read more about the training phase of the mice, please see our [behavioral paper](https://doi.org/10.7554/eLife.63711)

pipeline.png

## Experimental rig




All recordings were conducted in a standardised electrophysiology rig replicated across the 12 experimental labs who collected data. The rig comprises the instrumentation to conduct the behavioral task (reward valve, screen, wheel etc), three cameras and up to two Neuropixel probes each attached to a micromanipulator.


image (2).png

## The task


In the IBL decision making task a visual stimulus appears at the edge of the screen and mice must bring the stimulus to the screen centre by moving a wheel with their front two paws. If they successfully bring the stimulus to the screen centre, they recieve a reward of sugar water ; if however they move the wheel in the wrong direction until the stimulus reaches beyond the screen edge, a white noise tone is played. To initiate a trial, the mouse must hold the wheel still for a continous period between 0.4-0.7s ; they are then alerted of the start of the trial by the simultaneous presentation of the stimulus on the screen and a go cue tone. Mice have a maximum of 60s to make a decision before the trial times out and a white noise tone is played.

Varying contrasts of visual stimulus are shown throughout the session (100%, 25%, 12.5%, 6.25% and 0%). The probability of the stimulus appearing on the left or the right changes between blocks of trials. During 0% contrast trials (where no stimulus appears on the screen but a wheel response is required), the mice can use the inferred block structure to guide their decision.

# Overview of data

## Behavioural Data

The IBL rig contains hardware and software to control the behavioural task.

Behavioral events of interest that are extracted include:

*   the stimulus onset time
*   the mouse's choice
*   the feedback delivery time
*   the feedback type (reward vs error tone)
*   the probability of the stimulus being on the left
*   the wheel motion onset time. The continuous recording of the wheel is also extracted.



behavior.png

## Video data

The IBL rig contains three cameras:


*   a `left` camera and a `right` camera that record the mouses face, paws, and upper body from the side.
*   a `body` camera that records the mouses body in the holder from above.

The three recorded videos are passed through a DeepLabCut (DLC) pipeline to perform pose estimation. Tracked points include:

For each the `left` and `right` cameras:
*   the tongue
*   the paws
*   the pupil




For the `body` camera:
*   the tail

The motion energy of the whisker pads (`left` and `right` camera) and body (`body` camera ) are also computed.


Please refer to our [video white paper](https://figshare.com/articles/online_resource/Video_hardware_and_software_for_the_International_Brain_Laboratory/19694452) for more detailed information on the video processing pipeline

## Electrophysiology data

In a session, neural data from up to two Neuropixel probes was recorded.

A number of electrophysiology features are extracted from the neural data. These include among others,

*   root mean square of AP and LFP band
*   power spectrum of the AP and LFP band

The neural data are also spikesorted using pyKilosort to extract single neurons from the data.


Please refer to our [spikesorting white paper](https://figshare.com/articles/online_resource/Spike_sorting_pipeline_for_the_International_Brain_Laboratory/19705522) for more detailed information on the electrophysiology processing pipeline

## Metadata

Metadata associated with each experimental session is stored in a database (OpenAlyx). Relevant metadata is automatically uploaded to the database during different parts of the experimental pipeline. The collection of this metadata allows us to easily search the database to retrieve data from experiments matching certain criteria, for example

*   experimental sessions from particular mouse
*   experimental sessions with two Neuropixel probes
*   experimental sessions that targetted a particular brain area

# Data organisation



## Folder structure


All the data for a single experimental session is organised into a session folder that follows the following convention,

**`lab/Subjects/subject_name/date/session_number`**

For example ` mainenlab/Subjects/ZFM-01576/2020-12-01/001/`.

Within this session folder data is organised into a number of sub folder, e.g `alf` and `raw_video_data`.




```python
mainenlab/
 Subjects/
   ZFM-01576/
     2020-12-01/
       001/
         alf/
           probeXX/
              pykilosort/
         raw_ephys_data
           probeXX/
         raw_video_data
```

Data that are stored in the `raw_data` folders contain the unprocessed data from the different devices.

Data stored in the `alf` contains the procesed data and all data within this folder has been synced to a common clock

timestamps in alf seconds from experiment start

## File naming convention



IBL files follow the [AlyxFile](https://int-brain-lab.github.io/ONE/alf_intro.html) (ALF) naming convention. Each file has at least three parts, an **object**, an **attribute** and an **extension**. All files with the same object have the same number of rows and when an attribute of one file matches the object of another it indicates a cross reference. For example `spikes.clusters.npy` and `cluster.amps.npy`.

# Data Access

## ONE API





IBL data consists of physical files that are stored on a remote data server (Amazon S3 bucket) and metadata that are stored in a database (OpenAlyx). The [Open Neurophysiological Environment](https://int-brain-lab.github.io/ONE/index.html) (ONE) provides an api that communicates between the local user, OpenAlyx and the data server to allow you to search, download and load IBL data of your choice.
